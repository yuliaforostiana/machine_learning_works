**2. Арифметичні операції**

Створіть дві змінні: `var1` зі значенням 15, `var2` зі значенням 4.

Обчисліть та виведіть (кожен пункт оцінюється в 0.5 бала):
- Суму
- Різницю
- Добуток
- Ділення
- Цілочисельне ділення
- Залишок від ділення
- Піднесення до степеня


```python
var1 = 15
var2 = 4
```

```python
sum_var = var1 + var2 # sum
diff_var = var1 - var2 # difference
prod_var = var1 * var2 # product
div_var = var1 / var2 # division
int_div_var = var1 // var2 # integer division
rem_div_var = var1 % var2 # remainder from division
exp_var1 = var1 ** var2 # exponentiation
exp_var2 = var2 ** var1 # exponentiation
```

```python
print(f"The variable 1 equals {var1} and the variable 2 is {var2}.")
print(f"The sum of them equals {sum_var} while the difference is {diff_var}.")
print(f"The product of them equals {prod_var} while the difference is {div_var}. Besides integer division equals {int_div_var} while remainder from division is {rem_div_var}.")
print(f"If {var1} raised to the {var2} power then we get {exp_var1}. But if {var2} raised to the {var1}, we have {exp_var2}.")
```
