**1. Функція з параметром**

Створіть функцію `show_menu()`, яка приймає перелік страв, наприклад ['Піца', 'Паста', 'Салат'] і виводить меню ресторану в наступному вигляді:
```
=== МЕНЮ ===
1. Піца
2. Паста
3. Салат
=============
```

Викличте цю функцію на списку ["Суп грибний", "Котлета по-київськи", "Лимонад"].

````python
dishes_1 = ['Піца', 'Паста', 'Салат']

def show_menu(dishes):
    print('=== МЕНЮ ===')
    for i, dish in enumerate(dishes):
        print(f'{i+1}. {dish}')
    print('='*12)

show_menu(dishes_1)

dishes_2 = ["Суп грибний", "Котлета по-київськи", "Лимонад"]
show_menu(dishes_2)
````

**2. Функція з параметром**

Створіть функцію `print_square(n)`, яка виводить квадрат числа.

Наприклад: `print_square(5)` → "Квадрат числа 5 = 25"

Викличте її з різними числами.

````python
import random 

random.seed(77)
numbers = list(random.randint(1,200) for _ in range(15))
print('The list of numbers:', numbers)
````

````python
# спосіб 1
def print_square(n):
    for number in numbers:
        print(f'Квадрат числа {number} = {number ** 2}')

print_square(numbers)
````

````python
# спосіб 2
print_squares = list(map(lambda n: print(f'Квадрат числа {n} = {n ** 2}'), numbers))
````

**4. Функція пошуку максимумального значення**

Створіть функцію `find_max(numbers)`, яка:
- Приймає список чисел
- Повертає найбільше число в списку

Протестуйте на списку `[15, 42, 8, 23, 56, 19]`.


````python
list_num = [15, 42, 8, 23, 56, 19]
````

````python
# спосіб 1
def find_max(numbers):
    print('The list is', numbers)
    max_num = max(numbers)
    return max_num

result = find_max(list_num)
print('The max number of the list is', result)
````

````python
# спосіб 2
find_max = (lambda numbers: max(numbers)) (list_num)
print('The list is', list_num)
print('The max number of the list is', find_max)
````

**5. Функція з умовами**

Створіть функцію `check_grade(grade)`, яка:
- Приймає оцінку (число)
- Повертає "Здав" якщо grade >= 60
- Повертає "Не здав" якщо grade < 60

Протестуйте на різних оцінках.


````python
grade = 95

def check_grade(grade = 60):
    if grade >= 60:
        print("Здав")
    else:
        print("Не здав")

check_grade(grade)
check_grade()
check_grade(56)
````
