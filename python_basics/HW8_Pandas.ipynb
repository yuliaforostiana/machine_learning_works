**5. Статистика із групуванням**

Обчисліть (кожен пункт оцінюється в 0.5 бала):
- Середній рахунок за статтю (groupby `sex`)
- Середні чайові за днем тижня (groupby `day`)
- Скільки відвідувачів у кожен день

```python
print('Середній рахунок за статтю')
df.groupby('sex')['total_bill'].mean().round(2)
```

Загалом чоловіки мають вищий середній чек, ніж жінки.

```python
print('Середні чайові за днем тижня')
df.groupby('day')['tip'].mean().round(2)
```

Найвищий середній чек спостерігається у неділю, при цьому в п'ятницю найменший.

```python
print('Кількість відвідувачів по днях')
df.groupby('day')['total_bill'].count()
```

Найменша кількість відвідувачів спостерігається у п'ятницю, при цьому, скоріш за все, вони замовляють їжу низької-середньої вартості, оскільки саме в даний день спостерігається найменший середній чек. При цьому можемо припустити, що найвищий середній чек у неділю забезпечується придбанням дорожчої продукції, оскільки найбільша кількість відвідувачів спостерігається в суботу.
