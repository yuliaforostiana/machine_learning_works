**1. Створення масивів**

Створіть (кожен пункт оцінюється в 0.5 бала):
- Масив зі списку `[5, 15, 25, 35, 45]`
- Масив з нулів розміром 7
- Масив з одиниць розміром 6
- Масив з числами від 10 до 19 використовуючи `np.arange()`

```python
arr_list = np.array([5, 15, 25, 35, 45])
arr_zeros = np.zeros(shape = 7)
arr_ones = np.ones(shape = 6)
arr_num = np.arange(10,20)

print(f'Масив зі списку: {arr_list}')
print(f'Масив з нулів розміром 7: {arr_zeros}')
print(f'Масив з одиниць розміром 6: {arr_ones}')
print(f"Масив з числами від 10 до 19 використовуючи `np.arange()`: {arr_num}")
```

**2. Базові операції**

Створіть Numpy-масив зі значеннями [12, 24, 36, 48, 60].

Виконайте наступні операції над **оригінальним** масивом і запишіть у нову змінну результат (кожен пункт оцінюється в 0.5 бала):
- Помножте всі елементи на 2
- Додайте до всіх елементів 10
- Піднесіть всі елементи до квадрату
- Обчисліть квадратний корінь з усіх елементів

Виведіть результат виконання кожної операції.

```python
arr_1 = np.array([12, 24, 36, 48, 60])
print(f'Початковий масив: {arr_1}')

operation_1 = arr_1 * 2
print(f'Елементи масиву {arr_1} помножені на 2: {operation_1}')

operation_2 = arr_1 + 10
print(f'До кожного елементу масиву {arr_1} додано 10: {operation_2}')

operation_3_1 = arr_1 ** 2
print(f'Всі елементи масиву {arr_1} піднесені до квадрату за допомогою звичайного обчислення: {operation_3_1}')

operation_3_2 = np.square(arr_1)
print(f'Всі елементи масиву {arr_1} піднесені до квадрату за допомогою функції numpy: {operation_3_2}')

operation_4 = arr_1 ** (1/2)
print(f'Взято корінь квадратний з усіх елементів масиву {arr_1}: {operation_4.round(4)}')
```

**3. Статистичні функції**

Створіть Numpy-масив `data` зі значеннями [18, 27, 33, 41, 52, 68, 79].

Обчисліть використовуючи функціонал numpy (кожен пункт оцінюється в 0.5 бала):
- Середнє значення
- Мінімум
- Максимум
- Суму всіх елементів

```python
data = np.array([18, 27, 33, 41, 52, 68, 79])
print(f'Початковий масив: {data}')

data_mean = np.mean(data)
print(f'Середнє значення масиву {data} становить {data_mean.round(4)}')

data_min = np.min(data)
print(f'Мінімальне значення масиву {data} становить {data_min}')

data_max = np.max(data)
print(f'Максимальне значення масиву {data} становить {data_max}')

data_sum = np.sum(data)
print(f'Сума всіх елементів масиву {data} становить {data_sum}')
```

**4. Індексація**

Створіть Numpy-масив `arr` зі значеннями [15, 22, 35, 44, 58, 67].

Виведіть (кожен пункт оцінюється в 0.5 бала):
- Перший елемент
- Останній елемент
- Елементи з індексами 1 до 3
- Елементи більші за 40

```python
arr = np.array([15, 22, 35, 44, 58, 67])
print(f'Початковий масив: {arr}')

print(f'Першим елементом масиву {arr} є {arr[0]}')

print(f'Останнім елементом масиву {arr} є {arr[-1]}')

print(f'Елементи з індексами 1 до 3 масиву {arr} є {arr[1:4]}')

mask = arr > 40
print(f'Елементи масиву {arr}, які більші за 40: {arr[mask]}')
```

**5. Двовимірні масиви**

Задано матрицю 3x3.

Виконайте наступні дії з функціоналом Numpy (кожен пункт оцінюється в 0.5 бала):
- Виведіть форму (shape) матриці
- Виведіть перший рядок
- Виведіть другу колонку
- Обчисліть суму всіх елементів

```python
print('Працюємо з наступною матрицею:')
display(matrix)

print(f'Форма матриці: {matrix.shape}')

print(f'Перший рядок матриці є {matrix[0,:]}')

print(f'Друга колонка матриці є {matrix[:,1]}')

print(f'Сума всіх елементів матриці становить: {np.sum(matrix)}')

print(f'Середнє всіх елементів матриці становить: {np.mean(matrix)}')
```
